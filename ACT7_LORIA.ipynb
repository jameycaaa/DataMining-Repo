```python
import numpy as np
import pandas as pd 
# import seaborn as sns 
# import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap 
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('dataset\\terrorism.csv', encoding='ISO-8859-1')
df
```

# Crime Type Analysis
<h3>Filter for "Bombing" attacks after 2015 with at least 10 casualties.</h3>

```python
filtered_df1 = df[df['attacktype1_txt'] == 'Bombing/Explosion']
filtered_df1 = filtered_df1[filtered_df1['iyear'] >= 2015]
filtered_df1 = filtered_df1[filtered_df1['nkill'] >= 10]
print(f"Filtered dataset size: {len(filtered_df1)} rows")

filtered_df1.head(10)
aggregated_df1 = filtered_df1.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

print(aggregated_df1.head())
```

```python
filtered_df1 = df[(df['attacktype1_txt'] == 'Bombing/Explosion') & (df['iyear'] >= 2015) & (df['nkill'] >= 10)]

aggregated_df1 = filtered_df1.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

aggregated_df1['normalized_weight'] = (aggregated_df1['incident_count'] - aggregated_df1['incident_count'].min()) / \
                                       (aggregated_df1['incident_count'].max() - aggregated_df1['incident_count'].min())

heat_data1 = aggregated_df1[['latitude', 'longitude', 'normalized_weight']].values.tolist()

m1 = folium.Map(location=[0,0], zoom_start=2)

HeatMap(heat_data1).add_to(m1)

m1.save('bombing_heatmap.html')
m1
```
